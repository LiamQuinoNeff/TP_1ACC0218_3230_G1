# Traductor de PSeInt a Python

## Proyecto de Teoría de Compiladores - Traductor Automático

Un proyecto educativo para entender cómo funcionan los compiladores y traductores, construyendo un sistema que convierte código escrito en **PSeInt** (pseudocódigo didáctico) a **Python** (código ejecutable).

## Objetivo del Proyecto

Este notebook tiene como objetivo desarrollar un **traductor automático** que:

1. **Analiza** código escrito en PSeInt (pseudocódigo estructurado)
2. **Valida** la sintaxis mediante un analizador léxico y sintáctico
3. **Transforma** el árbol de análisis sintáctico (AST) en código Python equivalente
4. **Genera** código Python ejecutable que produce los mismos resultados

A través de este proyecto, exploraremos conceptos fundamentales de compiladores:
- **Análisis léxico**: tokenización del código fuente
- **Análisis sintáctico**: construcción del árbol de análisis sintáctico
- **Análisis semántico**: transformación a código destino
- **Generación de código**: creación de código ejecutable

## ¿Qué es PSeInt?

**PSeInt** es un lenguaje de pseudocódigo didáctico diseñado para enseñar programación de manera estructurada y con una sintaxis clara, independiente de cualquier lenguaje de programación específico.

### Características principales de PSeInt:

- **Sintaxis simple y clara**: Fácil de entender para principiantes
- **Estructuras de control**: Si/FinSi, Mientras, Para, Repetir/HastQue
- **Tipos de datos**: Entero, Real, Carácter, Lógico
- **Procedimientos y Funciones**: Reutilización de código
- **Entrada/Salida**: Leer (input) y Escribir (output)
- **Legibilidad**: Enfocado en la lógica, no en detalles de sintaxis

### Ejemplo simple en PSeInt:

```
Algoritmo Suma
	Definir a, b, resultado Como Entero
	Escribir "Ingrese el primer número:"
	Leer a
	Escribir "Ingrese el segundo número:"
	Leer b
	resultado <- a + b
	Escribir "La suma es: ", resultado
FinAlgoritmo
```

## ¿Por qué traducir PSeInt a Python?

### Puente entre educación y ejecución

PSeInt es excelente para **aprender** conceptos de programación, pero carece de:
- Ecosistema de librerías completo
- Capacidad de ejecutar código en diferentes plataformas fácilmente
- Integración con herramientas profesionales

**Python**, por otro lado, es:
- Un lenguaje versátil y poderoso
- Fácil de leer y escribir (similar a pseudocódigo)
- Ampliamente utilizado en la industria
- Con un ecosistema gigante de librerías

### Beneficios del traductor

✅ **Para estudiantes**: Transición natural de pseudocódigo a lenguaje de programación real  
✅ **Para educadores**: Herramienta para enseñar compiladores mediante un ejemplo práctico  
✅ **Para desarrolladores**: Automatización que ahorra tiempo en conversión manual de código

# Análisis de la gramática de PSeInt

### Introducción

PSeInt utiliza un pseudolenguaje sencillo y estructurado orientado a la enseñanza de algoritmos. Su gramática es relativamente fácil de procesar porque emplea palabras clave explícitas como FinSi, FinMientras o FinProceso, evitando depender de la indentación como ocurre en Python.

Para desarrollar un traductor de PSeInt a Python, el lenguaje puede dividirse en:

- Tokens
- Sentencias
- Expresiones
- Estructuras de control
- Funciones y subprocesos
---
### Estructura general del programa

Todo programa inicia con Proceso o Algoritmo y finaliza con FinProceso o FinAlgoritmo.

```
Proceso MiPrograma
    acciones;
FinProceso
```

Gramática simplificada:
```
programa ::= encabezado bloque fin

encabezado ::= ("Proceso" | "Algoritmo") IDENTIFICADOR
```
---
### Componentes principales de la gramática
#### 1. Declaraciones

Permiten definir variables y tipos de datos.
```
Definir x Como Entero;
Definir nombre Como Cadena;
```
Tipos básicos:
- Entero
- Real
- Logico
- Caracter
- Cadena

#### 2. Asignaciones

Usan el operador ```<-```.
```
x <- 5;
suma <- a + b;
```
Equivalencia en Python:
```
x = 5
suma = a + b
```
#### 3. Entrada y salida
##### Leer
```
Leer nombre;
```
↓
```
nombre = input()
```
##### Escribir
```
Escribir "Hola";
```
↓
```
print("Hola")
```
---
### Estructuras de control
##### Condicional ```Si```
```
Si edad >= 18 Entonces
    Escribir "Mayor";
Sino
    Escribir "Menor";
FinSi
```
Gramática simplificada:
```
si ::= "Si" expresion "Entonces"
          bloque
       ("Sino" bloque)?
       "FinSi"
```
##### Ciclo ```Mientras```
```
Mientras x < 10 Hacer
    x <- x + 1;
FinMientras
```
↓
```
while x < 10:
    x += 1
```
##### Ciclo ```Para```
```
Para i <- 1 Hasta 10 Hacer
    Escribir i;
FinPara
```
↓
```
for i in range(1, 11):
    print(i)
```
##### Ciclo ```Repetir```

Equivalente a un ```do-while```.
```
Repetir
    Leer x;
Hasta Que x > 0;
```
---
### Expresiones

Las expresiones incluyen operadores:

#### Aritméticos
```
+
-
*
/
^
MOD
```
#### Relacionales
```
>
<
>=
<=
=
<>
```
#### Lógicos
```Y``` ó ```&```

```|``` ó ```O```

```NO``` ó ```~```

---
### Jerarquía de operadores

El parser debe respetar la precedencia:

1. Paréntesis
2. Potencia
3. Multiplicación y división
4. Suma y resta
5. Operadores relacionales
6. Operadores lógicos
---
### Funciones y SubProcesos
#### Funciones
Retornan un valor.
```
Funcion resultado <- sumar(a, b)
    resultado <- a + b;
FinFuncion
```
#### SubProcesos
No retornan valores.
```
SubProceso saludar(nombre)
    Escribir "Hola ", nombre;
FinSubProceso
```

# Definición de la gramática en ANTLR4

A continuación se presenta la **gramática completa**. Respecto a la versión inicial (que solo cubría declaraciones, asignaciones y E/S) se completaron las partes pendientes:

### ✅ Estructuras de control
- **Condicional** `Si … Entonces … Sino … FinSi`
- **Bucle `Mientras … Hacer … FinMientras`** (equivale a `while`)
- **Bucle `Para … Hasta … [Con Paso …] Hacer … FinPara`** (equivale a `for`)
- **Bucle `Repetir … Hasta Que …`** (equivale a un `do-while`)

### ✅ Operadores relacionales y lógicos
Necesarios para escribir las condiciones de los `Si` y los bucles:
- Relacionales: `>` `<` `>=` `<=` `=` `<>`
- Lógicos: `Y` / `&`, `O` / `|`, `NO` / `~`

### ✅ Booleanos en pseudocódigo
Se agregó el tipo de literal **`BOOLEANO`** con los valores **`Verdadero`** y **`Falso`**, que el traductor convierte a `True` / `False` de Python.

### ✅ Funciones y subprocesos
Se completó la última parte pendiente de la gramática:
- **`Funcion ret <- nombre(args) … FinFuncion`** → `def nombre(args): … return ret`
- **`SubProceso nombre(args) … FinSubProceso`** → `def nombre(args): …` (sin retorno)
- **Llamadas**: como expresión (`x <- sumar(3, 4)`) o como sentencia (`saludar("Mundo")`)

```
Funcion resultado <- sumar(a, b)
    resultado <- a + b;
FinFuncion

SubProceso saludar(nombre)
    Escribir "Hola ", nombre;
FinSubProceso

Algoritmo Principal
    Definir s Como Entero
    s <- sumar(7, 5);
    saludar("Mundo");
FinAlgoritmo
```

### ✅ Jerarquía de precedencia
ANTLR resuelve la precedencia según el **orden de las alternativas** de la regla `expr` (la primera tiene mayor prioridad). La potencia `^` se declara asociativa por la derecha con `<assoc=right>`.


````python
%%writefile PSeInt.g4
grammar PSeInt;

// ============================================
// Estructura básica
// ============================================
// Un programa puede incluir funciones y subprocesos (antes o después
// del algoritmo principal), seguidos del algoritmo/proceso principal.
prog: subprograma* principal subprograma* EOF;

principal: encabezado bloque fin;

encabezado: ('Proceso' | 'Algoritmo') IDENTIFICADOR;
bloque: sentencia*;
fin: ('FinProceso' | 'FinAlgoritmo');

// ============================================
// Funciones y subprocesos
// ============================================

subprograma: funcion | subproceso;

// Funcion: retorna un valor (declarado como variable de retorno)
funcion
    : 'Funcion' IDENTIFICADOR '<-' IDENTIFICADOR '(' parametros? ')'
        bloque
      'FinFuncion'
    ;

// SubProceso / SubAlgoritmo: no retorna valor
subproceso
    : ('SubProceso' | 'SubAlgoritmo') IDENTIFICADOR '(' parametros? ')'
        bloque
      ('FinSubProceso' | 'FinSubAlgoritmo')
    ;

parametros: IDENTIFICADOR (',' IDENTIFICADOR)*;
argumentos: expr (',' expr)*;

// ============================================
// Sentencias
// ============================================

sentencia
    : declaracion
    | asignacion
    | lectura
    | escritura
    | condicional   // Si / Sino / FinSi
    | mientras      // Mientras / FinMientras
    | para          // Para / FinPara
    | repetir       // Repetir / Hasta Que
    | llamada       // invocación de un subproceso
    ;

// Declaraciones
declaracion: 'Definir' IDENTIFICADOR (',' IDENTIFICADOR)* 'Como' tipo ';'?;
tipo: 'Entero' | 'Real' | 'Logico' | 'Caracter' | 'Cadena';

// Asignaciones (el ';' final es opcional para mayor tolerancia)
asignacion: IDENTIFICADOR '<-' expr ';'?;

// Entrada/Salida
lectura: 'Leer' IDENTIFICADOR (',' IDENTIFICADOR)* ';'?;
escritura: ('Escribir' | 'Mostrar') listaEscritura ';'?;
listaEscritura: expr (',' expr)*;

// Llamada a un subproceso como sentencia
llamada: IDENTIFICADOR '(' argumentos? ')' ';'?;

// ============================================
// Estructuras de control
// ============================================

condicional
    : 'Si' expr 'Entonces'
        bloque
      ('Sino' bloque)?
      'FinSi'
    ;

mientras
    : 'Mientras' expr 'Hacer'
        bloque
      'FinMientras'
    ;

para
    : 'Para' IDENTIFICADOR '<-' expr 'Hasta' expr ('Con' 'Paso' expr)? 'Hacer'
        bloque
      'FinPara'
    ;

repetir
    : 'Repetir'
        bloque
      'Hasta' 'Que' expr ';'?
    ;

// ============================================
// Expresiones (con jerarquía de precedencia)
// ============================================
// De MAYOR a MENOR precedencia:
//   1. Paréntesis y llamadas a función
//   2. Potencia (^)               -> asociativa por la derecha
//   3. Multiplicación, división, módulo (*, /, MOD)
//   4. Suma, resta (+, -)
//   5. Relacionales (>, <, >=, <=, =, <>)
//   6. NO / ~  (negación lógica)
//   7. Y / &   (conjunción)
//   8. O / |   (disyunción)
expr
    : '(' expr ')'                                          # Parens
    | IDENTIFICADOR '(' argumentos? ')'                    # LlamadaFuncion
    | <assoc=right> expr '^' expr                          # Potencia
    | expr ('*' | '/' | 'MOD') expr                        # MulDiv
    | expr ('+' | '-') expr                                # AddSub
    | expr ('>' | '<' | '>=' | '<=' | '=' | '<>') expr     # Relacional
    | ('NO' | '~') expr                                    # NoLogico
    | expr ('Y' | '&') expr                                # YLogico
    | expr ('O' | '|') expr                                # OLogico
    | NUMERO                                                # Numero
    | BOOLEANO                                              # Booleano
    | IDENTIFICADOR                                         # Variable
    | CADENA                                                # String
    ;

// ============================================
// TOKENS
// ============================================

// Booleanos en pseudocódigo: Verdadero / Falso
// (debe ir ANTES de IDENTIFICADOR para que ANTLR los reconozca como literales)
BOOLEANO: 'Verdadero' | 'Falso';

NUMERO: [0-9]+ ('.' [0-9]+)?;
IDENTIFICADOR: [a-zA-Z_][a-zA-Z0-9_]*;
CADENA: '"' (~["\\\r\n])* '"';

WS: [ \t\r\n]+ -> skip;
COMENTARIO: '//' ~[\r\n]* -> skip;
````

````python
from antlr4 import *
from PSeIntLexer import PSeIntLexer
from PSeIntParser import PSeIntParser

def parse_pseint(code_text):
    """Función auxiliar para parsear código PSeint y mostrar el árbol"""
    input_stream = InputStream(code_text)
    lexer = PSeIntLexer(input_stream)
    token_stream = CommonTokenStream(lexer)
    parser = PSeIntParser(token_stream)
    tree = parser.prog()
    return tree, parser

print("✓ Librerías importadas y función de parsing lista")
````

# Generación de Código: el Visitor traductor

Hasta ahora solo hemos **mostrado** el árbol de análisis sintáctico (AST). El siguiente paso es **aplicarlo**: recorrer el AST y generar el código Python equivalente. Para esto usamos el patrón **Visitor** que ANTLR genera automáticamente (`PSeIntVisitor`).

## ¿Cómo funciona?

ANTLR crea una clase base `PSeIntVisitor` con un método `visitX` por cada regla / alternativa etiquetada de la gramática. Nosotros creamos una subclase, **`TraductorPython`**, que sobreescribe esos métodos para que **devuelvan o emitan código Python**:

| Construcción PSeInt | Método del Visitor | Python generado |
|---|---|---|
| `Si … Entonces … Sino … FinSi` | `visitCondicional` | `if …: / else:` |
| `Mientras … Hacer … FinMientras` | `visitMientras` | `while …:` |
| `Para i <- a Hasta b Hacer … FinPara` | `visitPara` | `for i in range(a, b + 1):` |
| `Repetir … Hasta Que c` | `visitRepetir` | `while True: … if c: break` |
| `Verdadero` / `Falso` | `visitBooleano` | `True` / `False` |
| `Y` / `O` / `NO` | `visitYLogico` / `visitOLogico` / `visitNoLogico` | `and` / `or` / `not` |
| `=` / `<>` / `MOD` / `^` | `visitRelacional` / `visitMulDiv` / `visitPotencia` | `==` / `!=` / `%` / `**` |

### Estrategia de traducción
- Las **expresiones** (`visitNumero`, `visitAddSub`, …) **devuelven un `str`** con el fragmento Python.
- Las **sentencias** (`visitAsignacion`, `visitCondicional`, …) **emiten líneas** en una lista, controlando la **indentación** con `self.nivel`, porque a diferencia de PSeInt (que usa `FinSi`, `FinMientras`), Python delimita los bloques por sangría.
- Las **lecturas** (`Leer`) se traducen usando el **tipo declarado** de la variable: `Entero → int(input())`, `Real → float(input())`.

````python
%%writefile TraductorPython.py
if __name__ is not None and "." in __name__:
    from .PSeIntVisitor import PSeIntVisitor
else:
    from PSeIntVisitor import PSeIntVisitor


class TraductorPython(PSeIntVisitor):

    def __init__(self):
        self.lineas = []
        self.nivel = 0
        self.tipos = {}

    def emitir(self, texto):
        self.lineas.append("    " * self.nivel + texto)

    def traducir(self, tree):
        self.lineas = []
        self.nivel = 0
        self.tipos = {}
        self.visit(tree)
        return "\n".join(self.lineas)

    def valor_inicial(self, tipo):
        return {"Entero": "0", "Real": "0.0", "Logico": "False",
                "Caracter": '""', "Cadena": '""'}.get(tipo, "None")

    def visitProg(self, ctx):
        for sub in ctx.subprograma():
            self.visit(sub)
            self.emitir("")
        self.visit(ctx.principal())

    def visitPrincipal(self, ctx):
        nombre = ctx.encabezado().IDENTIFICADOR().getText()
        self.emitir(f"# Algoritmo: {nombre}")
        self.visit(ctx.bloque())

    def visitBloque(self, ctx):
        sentencias = ctx.sentencia()
        if len(sentencias) == 0:
            self.emitir("pass")
            return
        for s in sentencias:
            self.visit(s)

    def visitFuncion(self, ctx):
        retorno = ctx.IDENTIFICADOR(0).getText()
        nombre = ctx.IDENTIFICADOR(1).getText()
        params = []
        if ctx.parametros() is not None:
            params = [t.getText() for t in ctx.parametros().IDENTIFICADOR()]
        self.emitir(f"# Funcion: {nombre}")
        self.emitir(f"def {nombre}({', '.join(params)}):")
        self.nivel += 1
        self.visit(ctx.bloque())
        self.emitir(f"return {retorno}")
        self.nivel -= 1

    def visitSubproceso(self, ctx):
        nombre = ctx.IDENTIFICADOR().getText()     # un solo IDENTIFICADOR (el nombre)
        params = []
        if ctx.parametros() is not None:
            params = [t.getText() for t in ctx.parametros().IDENTIFICADOR()]
        self.emitir(f"# SubProceso: {nombre}")
        self.emitir(f"def {nombre}({', '.join(params)}):")
        self.nivel += 1
        self.visit(ctx.bloque())
        self.nivel -= 1

    def visitLlamada(self, ctx):
        nombre = ctx.IDENTIFICADOR().getText()
        args = []
        if ctx.argumentos() is not None:
            args = [self.visit(e) for e in ctx.argumentos().expr()]
        self.emitir(f"{nombre}({', '.join(args)})")

    def visitDeclaracion(self, ctx):
        tipo = ctx.tipo().getText()
        for t in ctx.IDENTIFICADOR():
            nombre = t.getText()
            self.tipos[nombre] = tipo
            self.emitir(f"{nombre} = {self.valor_inicial(tipo)}")

    def visitAsignacion(self, ctx):
        nombre = ctx.IDENTIFICADOR().getText()
        valor = self.visit(ctx.expr())
        self.emitir(f"{nombre} = {valor}")

    def visitLectura(self, ctx):
        for t in ctx.IDENTIFICADOR():
            nombre = t.getText()
            tipo = self.tipos.get(nombre)
            if tipo == "Entero":
                self.emitir(f"{nombre} = int(input())")
            elif tipo == "Real":
                self.emitir(f"{nombre} = float(input())")
            else:
                self.emitir(f"{nombre} = input()")

    def visitEscritura(self, ctx):
        partes = [self.visit(e) for e in ctx.listaEscritura().expr()]
        self.emitir("print(" + ", ".join(partes) + ")")

    def visitCondicional(self, ctx):
        cond = self.visit(ctx.expr())
        self.emitir(f"if {cond}:")
        self.nivel += 1
        self.visit(ctx.bloque(0))
        self.nivel -= 1
        if ctx.bloque(1) is not None:
            self.emitir("else:")
            self.nivel += 1
            self.visit(ctx.bloque(1))
            self.nivel -= 1

    def visitMientras(self, ctx):
        cond = self.visit(ctx.expr())
        self.emitir(f"while {cond}:")
        self.nivel += 1
        self.visit(ctx.bloque())
        self.nivel -= 1

    def visitPara(self, ctx):
        var = ctx.IDENTIFICADOR().getText()
        exprs = ctx.expr()
        inicio = self.visit(exprs[0])
        fin = self.visit(exprs[1])
        if len(exprs) == 3:
            paso = self.visit(exprs[2])
            self.emitir(f"for {var} in range({inicio}, {fin} + 1, {paso}):")
        else:
            self.emitir(f"for {var} in range({inicio}, {fin} + 1):")
        self.nivel += 1
        self.visit(ctx.bloque())
        self.nivel -= 1

    def visitRepetir(self, ctx):
        self.emitir("while True:")
        self.nivel += 1
        self.visit(ctx.bloque())
        cond = self.visit(ctx.expr())
        self.emitir(f"if {cond}:")
        self.nivel += 1
        self.emitir("break")
        self.nivel -= 1
        self.nivel -= 1

    def visitParens(self, ctx):
        return f"({self.visit(ctx.expr())})"

    def visitLlamadaFuncion(self, ctx):
        nombre = ctx.IDENTIFICADOR().getText()
        args = []
        if ctx.argumentos() is not None:
            args = [self.visit(e) for e in ctx.argumentos().expr()]
        return f"{nombre}({', '.join(args)})"

    def visitPotencia(self, ctx):
        return f"{self.visit(ctx.expr(0))} ** {self.visit(ctx.expr(1))}"

    def visitMulDiv(self, ctx):
        op = {"*": "*", "/": "/", "MOD": "%"}[ctx.getChild(1).getText()]
        return f"{self.visit(ctx.expr(0))} {op} {self.visit(ctx.expr(1))}"

    def visitAddSub(self, ctx):
        op = ctx.getChild(1).getText()
        return f"{self.visit(ctx.expr(0))} {op} {self.visit(ctx.expr(1))}"

    def visitRelacional(self, ctx):
        op = {">": ">", "<": "<", ">=": ">=", "<=": "<=",
              "=": "==", "<>": "!="}[ctx.getChild(1).getText()]
        return f"{self.visit(ctx.expr(0))} {op} {self.visit(ctx.expr(1))}"

    def visitNoLogico(self, ctx):
        return f"not {self.visit(ctx.expr())}"

    def visitYLogico(self, ctx):
        return f"{self.visit(ctx.expr(0))} and {self.visit(ctx.expr(1))}"

    def visitOLogico(self, ctx):
        return f"{self.visit(ctx.expr(0))} or {self.visit(ctx.expr(1))}"

    def visitNumero(self, ctx):
        return ctx.NUMERO().getText()

    def visitBooleano(self, ctx):
        return "True" if ctx.BOOLEANO().getText() == "Verdadero" else "False"

    def visitVariable(self, ctx):
        return ctx.IDENTIFICADOR().getText()

    def visitString(self, ctx):
        return ctx.CADENA().getText()
````

````python
from TraductorPython import TraductorPython


def traducir_a_python(codigo_pseint, mostrar_arbol=False):
    input_stream = InputStream(codigo_pseint)
    lexer = PSeIntLexer(input_stream)
    token_stream = CommonTokenStream(lexer)
    parser = PSeIntParser(token_stream)
    tree = parser.prog()

    if parser.getNumberOfSyntaxErrors() > 0:
        raise SyntaxError("El codigo PSeInt tiene errores de sintaxis")

    if mostrar_arbol:
        print("AST:", tree.toStringTree(recog=parser), "\n")

    return TraductorPython().traducir(tree)


def demo(codigo_pseint, ejecutar=True, entradas=None):
    print("PSeInt (entrada):")
    print(codigo_pseint.strip())
    print("\nPython (traduccion generada):")
    print("-" * 60)
    codigo_python = traducir_a_python(codigo_pseint)
    print(codigo_python)
    if ejecutar:
        print("-" * 60)
        print("Salida al ejecutar el Python generado:")
        if entradas is not None:
            it = iter(entradas)
            def _input(*args):
                v = str(next(it))
                print(v)
                return v
            entorno = {"input": _input}
        else:
            entorno = {}
        exec(compile(codigo_python, "<generado>", "exec"), entorno)


print("Traductor listo: usa traducir_a_python(codigo) o demo(codigo, entradas=[...])")
````

## Conclusiones

Con la gramática completa y el Visitor traductor, el proyecto cubre el flujo completo de un compilador/traductor:

1. **Análisis léxico** — el lexer de ANTLR convierte el texto PSeInt en tokens (incluidos `Verdadero`, `Falso`, operadores relacionales/lógicos y las palabras clave de funciones).
2. **Análisis sintáctico** — el parser construye el AST reconociendo condicionales, los tres tipos de bucles, funciones/subprocesos y las expresiones con su jerarquía de precedencia.
3. **Análisis semántico / generación de código** — `TraductorPython` (Visitor) recorre el AST y **emite código Python equivalente**, gestionando la indentación de los bloques y el retorno de las funciones.
4. **Ejecución** — el Python generado se ejecuta y produce el resultado esperado.

# Traductor de PSeInt a Python

## Proyecto de Teoría de Compiladores - Traductor Automático

Un proyecto educativo para entender cómo funcionan los compiladores y traductores, construyendo un sistema que convierte código escrito en **PSeInt** (pseudocódigo didáctico) a **Python** (código ejecutable).

## Objetivo del Proyecto

Este notebook tiene como objetivo desarrollar un **traductor automático** que:

1. **Analiza** código escrito en PSeInt (pseudocódigo estructurado)
2. **Valida** la sintaxis mediante un analizador léxico y sintáctico
3. **Transforma** el árbol de análisis sintáctico (AST) en código Python equivalente
4. **Genera** código Python ejecutable que produce los mismos resultados

A través de este proyecto, exploraremos conceptos fundamentales de compiladores:
- **Análisis léxico**: tokenización del código fuente
- **Análisis sintáctico**: construcción del árbol de análisis sintáctico
- **Análisis semántico**: transformación a código destino
- **Generación de código**: creación de código ejecutable

## ¿Qué es PSeInt?

**PSeInt** es un lenguaje de pseudocódigo didáctico diseñado para enseñar programación de manera estructurada y con una sintaxis clara, independiente de cualquier lenguaje de programación específico.

### Características principales de PSeInt:

- **Sintaxis simple y clara**: Fácil de entender para principiantes
- **Estructuras de control**: Si/FinSi, Mientras, Para, Repetir/HastQue
- **Tipos de datos**: Entero, Real, Carácter, Lógico
- **Procedimientos y Funciones**: Reutilización de código
- **Entrada/Salida**: Leer (input) y Escribir (output)
- **Legibilidad**: Enfocado en la lógica, no en detalles de sintaxis

### Ejemplo simple en PSeInt:

```
Algoritmo Suma
	Definir a, b, resultado Como Entero
	Escribir "Ingrese el primer número:"
	Leer a
	Escribir "Ingrese el segundo número:"
	Leer b
	resultado <- a + b
	Escribir "La suma es: ", resultado
FinAlgoritmo
```

## ¿Por qué traducir PSeInt a Python?

### Puente entre educación y ejecución

PSeInt es excelente para **aprender** conceptos de programación, pero carece de:
- Ecosistema de librerías completo
- Capacidad de ejecutar código en diferentes plataformas fácilmente
- Integración con herramientas profesionales

**Python**, por otro lado, es:
- Un lenguaje versátil y poderoso
- Fácil de leer y escribir (similar a pseudocódigo)
- Ampliamente utilizado en la industria
- Con un ecosistema gigante de librerías

### Beneficios del traductor

✅ **Para estudiantes**: Transición natural de pseudocódigo a lenguaje de programación real  
✅ **Para educadores**: Herramienta para enseñar compiladores mediante un ejemplo práctico  
✅ **Para desarrolladores**: Automatización que ahorra tiempo en conversión manual de código  

---

### Comparación: PSeInt vs Python

| Aspecto | PSeInt | Python |
|--------|--------|--------|
| **Propósito** | Educativo, pseudocódigo | General, producción |
| **Curva de aprendizaje** | Muy baja | Baja |
| **Ejecución** | En intérprete PSeInt | Directamente en máquina |
| **Librerías** | Limitadas | Enormes |
| **Sintaxis** | Muy legible, española | Clara, inglés |

# Análisis de la gramática de PSeInt

### Introducción

PSeInt utiliza un pseudolenguaje sencillo y estructurado orientado a la enseñanza de algoritmos. Su gramática es relativamente fácil de procesar porque emplea palabras clave explícitas como FinSi, FinMientras o FinProceso, evitando depender de la indentación como ocurre en Python.

Para desarrollar un traductor de PSeInt a Python, el lenguaje puede dividirse en:

- Tokens
- Sentencias
- Expresiones
- Estructuras de control
- Funciones y subprocesos
---
### Estructura general del programa

Todo programa inicia con Proceso o Algoritmo y finaliza con FinProceso o FinAlgoritmo.

```
Proceso MiPrograma
    acciones;
FinProceso
```

Gramática simplificada:
```
programa ::= encabezado bloque fin

encabezado ::= ("Proceso" | "Algoritmo") IDENTIFICADOR
```
---
### Componentes principales de la gramática
#### 1. Declaraciones

Permiten definir variables y tipos de datos.
```
Definir x Como Entero;
Definir nombre Como Cadena;
```
Tipos básicos:
- Entero
- Real
- Logico
- Caracter
- Cadena

#### 2. Asignaciones

Usan el operador ```<-```.
```
x <- 5;
suma <- a + b;
```
Equivalencia en Python:
```
x = 5
suma = a + b
```
#### 3. Entrada y salida
##### Leer
```
Leer nombre;
```
↓
```
nombre = input()
```
##### Escribir
```
Escribir "Hola";
```
↓
```
print("Hola")
```
---
### Estructuras de control
##### Condicional ```Si```
```
Si edad >= 18 Entonces
    Escribir "Mayor";
Sino
    Escribir "Menor";
FinSi
```
Gramática simplificada:
```
si ::= "Si" expresion "Entonces"
          bloque
       ("Sino" bloque)?
       "FinSi"
```
##### Ciclo ```Mientras```
```
Mientras x < 10 Hacer
    x <- x + 1;
FinMientras
```
↓
```
while x < 10:
    x += 1
```
##### Ciclo ```Para```
```
Para i <- 1 Hasta 10 Hacer
    Escribir i;
FinPara
```
↓
```
for i in range(1, 11):
    print(i)
```
##### Ciclo ```Repetir```

Equivalente a un ```do-while```.
```
Repetir
    Leer x;
Hasta Que x > 0;
```
---
### Expresiones

Las expresiones incluyen operadores:

#### Aritméticos
```
+
-
*
/
^
MOD
```
#### Relacionales
```
>
<
>=
<=
=
<>
```
#### Lógicos
```Y``` ó ```&```

```|``` ó ```O```

```NO``` ó ```~```

---
### Jerarquía de operadores

El parser debe respetar la precedencia:

1. Paréntesis
2. Potencia
3. Multiplicación y división
4. Suma y resta
5. Operadores relacionales
6. Operadores lógicos
---
### Funciones y SubProcesos
#### Funciones
Retornan un valor.
```
Funcion resultado <- sumar(a, b)
    resultado <- a + b;
FinFuncion
```
#### SubProcesos
No retornan valores.
```
SubProceso saludar(nombre)
    Escribir "Hola ", nombre;
FinSubProceso
```

# Definición de la gramática en ANTLR4

````python
%%writefile PSeInt.g4
grammar PSeInt;

// ============================================
// Estructura básica, declaraciones y E/S
// ============================================

prog: encabezado bloque fin EOF;

encabezado: ('Proceso' | 'Algoritmo') IDENTIFICADOR;
bloque: sentencia*;
fin: ('FinProceso' | 'FinAlgoritmo');

sentencia
    : declaracion
    | asignacion
    | lectura
    | escritura
    // Pendiente: agregar Si/FinSi (condicional)
    // | condicional
    // Pendiente: agregar Mientras, Para, Repetir (bucles)
    // | mientras
    // | para
    // | repetir
    // Pendiente: agregar procedimientos y funciones
    // | subproceso
    // | funcion
    ;

// Declaraciones
declaracion: 'Definir' IDENTIFICADOR (',' IDENTIFICADOR)* 'Como' tipo;
tipo: 'Entero' | 'Real' | 'Logico' | 'Caracter' | 'Cadena';

// Asignaciones
asignacion: IDENTIFICADOR '<-' expr ';';

// Entrada/Salida
lectura: 'Leer' IDENTIFICADOR (',' IDENTIFICADOR)* ';';
escritura: 'Escribir' listaEscritura ';';
listaEscritura: expr (',' expr)*;

// Expresiones (aritmética básica)
expr
    : expr ('*' | '/' | 'MOD') expr      # MulDiv
    | expr ('+' | '-') expr              # AddSub
    | expr '^' expr                      # Potencia
    | '(' expr ')'                       # Parens
    | NUMERO                             # Numero
    | IDENTIFICADOR                      # Variable
    | CADENA                             # String
    ;

// ============================================
// TOKENS
// ============================================

NUMERO: [0-9]+ ('.' [0-9]+)?;
IDENTIFICADOR: [a-zA-Z_][a-zA-Z0-9_]*;
CADENA: '"' (~["\\\r\n])* '"';

WS: [ \t\r\n]+ -> skip;
COMENTARIO: '//' ~[\r\n]* -> skip;
````

### Pruebas de Parsing y Ejemplos Progresivos

Ahora verificaremos que la gramática reconoce correctamente diferentes construcciones de PSeInt, mostrando el árbol de análisis sintáctico (AST) para cada caso.

````python
!java -cp .:antlr-4.13.2-complete.jar org.antlr.v4.Tool PSeInt.g4 -Dlanguage=Python3 -visitor

print("✓ Gramática compilada exitosamente")
print("✓ Archivos generados: PSeIntLexer.py, PSeIntParser.py, PSeIntVisitor.py")
````

````python
from antlr4 import *
from PSeIntLexer import PSeIntLexer
from PSeIntParser import PSeIntParser

def parse_pseint(code_text):
    """Función auxiliar para parsear código PSeint y mostrar el árbol"""
    input_stream = InputStream(code_text)
    lexer = PSeIntLexer(input_stream)
    token_stream = CommonTokenStream(lexer)
    parser = PSeIntParser(token_stream)
    tree = parser.prog()
    return tree, parser

print("✓ Librerías importadas y función de parsing lista")
````

### Ejemplo 1: Declaración y Asignación Simple

Probamos la declaración de variables y asignación básica.

````python
ejemplo1 = """
Algoritmo Suma
Definir x, y Como Entero
x <- 10;
y <- 20;
FinAlgoritmo
"""

tree1, parser1 = parse_pseint(ejemplo1)
print("Ejemplo 1: Declaración y Asignación")
print("-" * 50)
print(tree1.toStringTree(recog=parser1))
````

### Ejemplo 2: Entrada y Salida

Probamos las operaciones de lectura (Leer) y escritura (Escribir).

````python
ejemplo2 = """
Proceso Saludo
Definir nombre Como Cadena
Escribir "Ingrese su nombre:";
Leer nombre;
Escribir "Hola, ", nombre;
FinProceso
"""

tree2, parser2 = parse_pseint(ejemplo2)
print("Ejemplo 2: Entrada y Salida")
print("-" * 50)
print(tree2.toStringTree(recog=parser2))
````

### Ejemplo 3: Expresiones Aritméticas

Probamos operadores aritméticos con precedencia correcta.

````python
ejemplo3 = """
Algoritmo Operaciones
Definir a, b, c, resultado Como Entero
a <- 5;
b <- 3;
c <- 2;
resultado <- a + b * c;
resultado <- a ^ 2 + b / c;
resultado <- (a + b) * c - a;
FinAlgoritmo
"""

tree3, parser3 = parse_pseint(ejemplo3)
print("Ejemplo 3: Expresiones Aritméticas")
print("-" * 50)
print(tree3.toStringTree(recog=parser3))
````

### Ejemplo 4: Caso Complejo (Integración Total)

Combinamos declaraciones, lectura, expresiones y escritura en un programa integrado.

````python
ejemplo4 = """
Algoritmo CalculadoraSimple
Definir num1, num2, suma, resta, multiplicacion Como Entero
Escribir "Ingrese el primer número:";
Leer num1;
Escribir "Ingrese el segundo número:";
Leer num2;
suma <- num1 + num2;
resta <- num1 - num2;
multiplicacion <- num1 * num2;
Escribir "Suma: ", suma;
Escribir "Resta: ", resta;
Escribir "Multiplicación: ", multiplicacion;
FinAlgoritmo
"""

tree4, parser4 = parse_pseint(ejemplo4)
print("Ejemplo 4: Caso Complejo (Integración Total)")
print("-" * 50)
print(tree4.toStringTree(recog=parser4))
print("\n✓ Todos los ejemplos se parsearon correctamente!")
print("✓ La gramática está funcionando correctamente.")
````
